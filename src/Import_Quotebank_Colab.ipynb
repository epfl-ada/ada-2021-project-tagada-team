```python
#Will allow to extract the domain name from url links
def get_domain(url):
    res = get_tld(url, as_object=True)
    return res.tld
```

## Create a json file
We keep only the quotations from Pop Francis and add two attribute value pairs : one for the count of words in the quotation and one for the url domains.

```python
from tld import get_tld 

path_to_files = ['/content/drive/MyDrive/Quotebank/quotes-2015.json.bz2', '/content/drive/MyDrive/Quotebank/quotes-2016.json.bz2', '/content/drive/MyDrive/Quotebank/quotes-2017.json.bz2', '/content/drive/MyDrive/Quotebank/quotes-2018.json.bz2', '/content/drive/MyDrive/Quotebank/quotes-2019.json.bz2', '/content/drive/MyDrive/Quotebank/quotes-2020.json.bz2' ]
path_to_out = '/content/drive/MyDrive/Pope-quotes-total.json.bz2'

with bz2.open(path_to_out, 'wb') as d_file:
    for i, path in enumerate(path_to_files):
        with bz2.open(path, 'rb') as s_file:
        
            for instance in s_file:
                instance = json.loads(instance) # loading a sample
                speaker = instance['speaker']

                #We keep all citations from the pope (Quotebank also considers Jorge Mario Bergoglio's citations as the Pope Francis citations)
                if speaker == 'Pope Francis':

                    #We add the domain names to the json file
                    urls = instance['urls'] # extracting list of links
                    domains = []
                    for url in urls:
                        tld = get_domain(url)
                        domains.append(tld)
                    instance['domains'] = domains 

                    #We add the number of words of the citation to the json file 
                    instance['nbr_words'] = len(re.findall(r'\w+', instance['quotation']))

                    d_file.write((json.dumps(instance)+'\n').encode('utf-8')) # writing in the new file

                
```

## Total number of quotes in our json file

```python
path_to_file = '/content/drive/MyDrive/Pope-quotes-total.json.bz2'
counter = 0

with bz2.open(path_to_file, 'rb') as s_file:
    for instance in s_file:
        counter = counter + 1

print(counter)
        
```
